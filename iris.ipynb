```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
# Load the data
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.columns
```

```python
#Separate features and target
data = df.values
X = data[:,1:5]
Y = data[:,5]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the data to train and test the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(y_test)
```

```python
#Training the model using Logistic Regression
from sklearn.linear_model import LogisticRegression#Creating instance for Logistic Regression model
model_LR = LogisticRegression()
#Training the model with training dataset
model_LR.fit(X_train,y_train)
```

```python
#Prediction from the model
prediction = model_LR.predict(X_test)
```

```python
#Comparing the predicted ones with the actual values
for i in range(len(prediction)):
  print(y_test[i], prediction[i])
```

```python
#Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction)*100)
```

```python
# A detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```

```python
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])
#Prediction of the species from the input vector
prediction = model_LR.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```
